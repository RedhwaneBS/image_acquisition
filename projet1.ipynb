```python
import matplotlib.pyplot as plt
import numpy as np
from skimage.io import imread,imsave,imshow
from skimage.color import rgb2hsv
from matplotlib.pyplot import cm

%matplotlib inline
```

```python
# Read the image
watermark = imread('watermark.png').astype('int16')
im = imread('etretat.jpg').astype('int16')

plt.imshow(watermark, cmap = cm.gray)
plt.show()
```

```python


def get_middle_point(im):
    """
    Returns the middle point of the image
    """
    return (im.shape[0]//2, im.shape[1]//2)

def add_watermark(image, watermark,opacity,center=get_middle_point(im),inversion=False):
    im2 = image.copy()
    watermark_shape = watermark.shape
    bool_light = 1
    hsv_image = rgb2hsv(image)
    if opacity > 100 or opacity < 0:
        raise ValueError("Opacity must be between 0 and 100")
    elif center[0] + watermark_shape[0] > im.shape[0] or center[1] + watermark_shape[1] > im.shape[1]:
        raise ValueError("Watermark is too big for the image")
    elif center[0] > im.shape[0] or center[1] > im.shape[1] or center[0] < 0 or center[1] < 0:
        raise ValueError("Center is out of bounds")
        
    for y in range(watermark_shape[0]):
        for x in range(watermark_shape[1]):
            if watermark[y,x] != 0:
                if inversion:
                    if hsv_image[center[0]+ y, center[1] + x,2]*100 > 0.5:
                        bool_light = 1
                    else:
                        bool_light = -1
                im2[center[0]+ y, center[1] + x] -= opacity  * watermark[y,x] * bool_light // 100
    return im2


```
